# Netflix Data Analysis

## Project Goal
This project explores the Netflix Titles dataset to understand trends in content distribution, production countries, and release patterns.

## Dataset
Netflix Titles Dataset (8807 records)

## Questions
1. Are there more Movies or TV Shows on Netflix?
2. Which countries produce the most Netflix content?
3. How has Netflix content changed over time?
4. What are the most common genres on Netflix?

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```
df = pd.read_csv("netflix_titles.csv")

df.head()
```

```
df.shape
```

```
df.columns
```

```
df.info()
```

```
df.isnull().sum()
```

```
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Unknown")
```

```
df['type'].value_counts()
```

```
sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")

plt.show()
```

### Insight

Movies dominate Netflix content compared to TV Shows.

This suggests that Netflix still focuses heavily on movie distribution while gradually expanding its TV series library.

```
top_countries = df['country'].value_counts().head(10)

top_countries
```

```
top_countries.plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
```

### Insight

The United States produces the largest amount of Netflix content.

Other major contributors include India and the United Kingdom, showing Netflix's strong global production network.

```
content_by_year = df['release_year'].value_counts().sort_index()

content_by_year.head()
```

```
content_by_year.plot()

plt.title("Netflix Content Released Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

### Insight

Netflix content production increased rapidly after 2015.

This reflects the rapid growth of streaming platforms and Netflix's investment in original content.

```
df['rating'].value_counts().head(10)
```

```
df['rating'].value_counts().head(10).plot(kind='bar')

plt.title("Top Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

### Insight

The most common ratings include TV-MA and TV-14, indicating that a large portion of Netflix content targets mature audiences.

```
df['listed_in'].value_counts().head(10)
```

```
df['listed_in'].value_counts().head(10).plot(kind='barh')

plt.title("Top Netflix Genres")
plt.xlabel("Number of Titles")

plt.show()
```

# Conclusion

Key findings from the analysis:

1. Movies dominate Netflix content compared to TV Shows.
2. The United States produces the most Netflix titles.
3. Netflix content production increased significantly after 2015.
4. Mature audience ratings such as TV-MA are very common.
5. Drama and international content are among the most popular genres.

This analysis highlights how Netflix expanded its content library globally over time.
